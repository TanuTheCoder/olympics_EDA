```python
# Distribution by Gender
sns.countplot(data=df,x="Sex")
plt.title("Distribution by Gender ")
plt.show()
```

```python
# Age Distribution
sns.histplot(data=df,x="Age",bins=10,kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
#Height Distribution
sns.histplot(data=df,x="Height",bins=20,kde=True)
plt.title("Height Distribution")
plt.show()
```

```python
sns.histplot(data=df,x="Weight",bins=20,kde=True)
plt.title("Weight Distribution")
plt.show()
```

```python
# Medal Distribution
sns.countplot(data=df,x="Medal")
plt.title("Medals Distribution ")
plt.show()
```

```python
# Medals Distribution over the Years
sns.countplot(data=df,x="Year",hue="Medal")
plt.title("Medals Distribution over the Years")
plt.xticks(rotation=60)
plt.show()
```

```python
# Average age of participant
year_avg_age=df.groupby('Year')['Age'].mean()
print(year_avg_age)
```

```python
# Sport Median Height
Sport_Median_Height=df.groupby('Sport')['Height'].median()
print(Sport_Median_Height)
```

```python
year_participant_count=df.groupby('Year')['ID'].nunique()
year_participant_count.plot(kind='line',marker='o')
plt.title('Number of Participants over the years')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.show()
```

```python
gender_counts = df['Sex'].value_counts()
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Distribution of Athletes')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Analyzing Performance Trends
medal_counts = df['Medal'].value_counts()
medal_counts.plot(kind='bar', color=['gold', 'silver', 'brown'])
plt.title('Medal Distribution')
plt.xlabel('Medal Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# height VS weight analysis
sns.scatterplot(x=df['Height'], y=df['Weight'], alpha=0.5)
plt.title('Height vs. Weight of Athletes')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
# most successful countries
top_countries = df['NOC'].value_counts()
top_countries.plot(kind='bar', color='green')
plt.title('Top 10 Countries with Most Athletes')
plt.xlabel('Country')
plt.ylabel('Number of Athletes')
plt.xticks(rotation=90)
plt.show()
```

```python
# Sports with the Most Athletes

top_sports = df['Sport'].value_counts().head(10)
top_sports.plot(kind='barh', color='purple')
plt.title('Top 10 Sports with Most Athletes')
plt.xlabel('Number of Athletes')
plt.ylabel('Sport')
plt.show()
```

```python
# Evolution of Olympic Events Over Time

event_count = df.groupby('Year')['Event'].nunique()
event_count.plot(kind='line', marker='o', color='red')
plt.title('Number of Unique Events Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Events')
plt.show()
```

```python
# Conclusions from the Olympic Dataset EDA
Based on the exploratory data analysis (EDA) performed on the Olympic dataset, we can derive the following key insights:
1. Growth in Olympic Participation Over Time:
The number of participants has increased significantly over the years, especially after the 1900s.
The trend suggests a rise in global interest and inclusion of more sports.
2. Gender Representation:
Historically, male athletes dominated the Olympics, but female participation has steadily increased.
Modern Olympic events have a much better gender balance compared to earlier years.
3. Variations in Athlete Heights Across Sports:
Certain sports have specific height trends.
Volleyball and Basketball: Taller athletes dominate.
Weightlifting and Gymnastics: Athletes tend to be shorter.
Archery and Wrestling: Height doesn’t have a strong influence.    
4. Correlation Between Height and Weight:
There is a positive correlation between height and weight, which is expected.
Outliers exist where athletes with unusually high or low weight appear in specific sports.
5. Most Popular Sports:
Certain sports have consistently high participation, including Athletics, Swimming, and Gymnastics.
New sports have been introduced over time, influencing participation trends.
6. Evolution of Olympic Events:
The number of unique Olympic events has increased, reflecting the diversification of the Olympics.
Some historical sports like Tug-Of-War were discontinued.
7. Country-Wise Performance Trends:
Certain countries dominate Olympic participation and medals, like the USA, China, and Russia.
Some smaller nations have fewer participants but excel in niche sports.
8. Medal Distribution:
Gold, Silver, and Bronze medals are distributed unevenly, with top-performing countries winning the majority.
Athletes specializing in certain events tend to dominate in their respective fields.
```

```python
# Final Takeaways:
The Olympics have evolved significantly, growing in participation, sports diversity, and gender inclusivity.
Athletes’ physical attributes vary widely based on their sport.
Historical trends suggest that participation will continue to grow, with new sports being introduced.
Certain countries maintain dominance, but smaller nations occasionally make breakthroughs in specific sports.
```
